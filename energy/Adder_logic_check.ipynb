```python
sum_df = [hoge for hoge in df.columns if ('SUM' in hoge)]
print(sum_df)
```

```python
result_bit_seq =[]
sum_df = [hoge for hoge in df.columns if ('SUM' in hoge)]
for col in sum_df:
    tmp = ['1' if hoge>0 else '0' for hoge in df[col].values[50::200] ]
    result_bit_seq.append(tmp)
```

```python
fooo= np.array(result_bit_seq).T
for item in fooo:
    bin_str = ''.join(item)
    dec_num = int(bin_str,2)
    print(f'{dec_num}\t\t{bin_str}')
```
