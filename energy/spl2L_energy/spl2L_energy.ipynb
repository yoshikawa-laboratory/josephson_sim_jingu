```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("/mnt/CAD_lib/home/yoshilab/ttanaka/shared/energy_josim_v263_04_22_14_10_21.csv",index_col = [0,1],header = None)
columns = [
    "bfr_-2_l","bfr_-2_r","bfr_-1_l","bfr_-1_r",
    "spl2L_l","spl2L_r",
    "bfr_+1_x_l","bfr_+1_x_r","bfr_+1_y_l","bfr_+1_y_r",
    "bfr_+2_x_l","bfr_+2_x_r","bfr_+2_y_l","bfr_+2_y_r",
 ]
df.index.names = ["input","freq"]
df.columns = columns
df = df/200
df["spl2L_l"] = df["spl2L_l"]*2
df["spl2L_r"] = df["spl2L_r"]*2
```

```python
df
```

```python
input_pat = '0b1'
see_comb = ''
foo = [x for x in df.columns if see_comb in x ]
#foo = df.columns
df2 = df.loc[input_pat,foo]
plt.figure(figsize=(12,8))
plt.plot(df2)
plt.legend(df2.columns)
plt.xscale("log")
plt.yscale("log")
plt.ylim(1e-24,1e-18)
plt.title(f'maj energy @{input_pat}')

```
